```python
import os
import fnmatch
import pandas as pd
import csv
import numpy as np
from pathlib import Path
from openpyxl import load_workbook
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")
```

```python
# white background 
sns.set(style="ticks", context="talk")
plt.rcParams.update({"grid.linewidth":0.5, "grid.alpha":0.8})
plt.style.use('default')
```

```python
# dark background
sns.set(style="darkgrid") #, context="talk")
sns.set(style="ticks", context="talk")
plt.rcParams.update({"grid.linewidth":0.5, "grid.alpha":0.8})
plt.style.use("dark_background")
```

```python
# Definitions

analysis_path = os.path.dirname(os.getcwd())

regions = ['accumbens', 'cerebellum', 'mPFC', 'OFC', 'frontal_cortex', 'hippocampus', 'striatum', 'thalamus', 'visual_cortex', 'whole_brain']


def group_allocation (row):
    if row['Treatment_Group_#'] == 1:
        return 'Veh'
    if row['Treatment_Group_#'] == 2:
        return 'Har'
    if row['Treatment_Group_#'] == 3:
        return 'DMT'
    if row['Treatment_Group_#'] == 4:
        return 'Har + DMT'
```

```python
df = pd.read_excel(f'{analysis_path}/02_raw_processed/static_average_scans/SUV_values/02_Hemispheres_averaged/01b_activity_in_relevant_regions_WB_uptake_normalized.xlsx', header = 0, engine = 'openpyxl')

df['Treatment_Group'] = df.apply(lambda row: group_allocation(row), axis=1)
```

## Whole brain uptake normalized data

```python
sns.color_palette("flare_r")
```

```python
f, axs = plt.subplots(3, 3, figsize=(20, 20), gridspec_kw=dict(width_ratios=[1, 1, 1]))
g = sns.boxplot(x='Treatment_Group', y='accumbens', data=df, showmeans=True, meanprops={"marker": "+",
                       "markeredgecolor": "black",
                       "markersize": "10"}, ax=axs[0,0], saturation=0.75, width=0.4, palette='flare_r')
g.set(ylim=(3.5, 8.5))


g = sns.boxplot(x='Treatment_Group', y='cerebellum', data=df, showmeans=True, meanprops={"marker": "+",
                       "markeredgecolor": "black",
                       "markersize": "10"},ax=axs[0,1], saturation=0.75, width=0.4, palette='flare_r') #boxprops={'alpha': 0.6}
g.set(ylim=(3.5, 8.5))


g = sns.boxplot(x='Treatment_Group', y='frontal_cortex', data=df, showmeans=True, meanprops={"marker": "+",
                       "markeredgecolor": "black",
                       "markersize": "10"},ax=axs[0,2], saturation=0.75, width=0.4, palette='flare_r')
g.set(ylim=(3.5, 8.5))

g= sns.boxplot(x='Treatment_Group', y='hippocampus', data=df, showmeans=True, meanprops={"marker": "+",
                       "markeredgecolor": "black",
                       "markersize": "10"},ax=axs[1,0], saturation=0.75, width=0.4, palette='flare_r')
g.set(ylim=(3.5, 8.5))

g = sns.boxplot(x='Treatment_Group', y='striatum', data=df, showmeans=True, meanprops={"marker": "+",
                       "markeredgecolor": "black",
                       "markersize": "10"},ax=axs[1,1], saturation=0.75, width=0.4, palette='flare_r')
g.set(ylim=(3.5, 8.5))

g = sns.boxplot(x='Treatment_Group', y='thalamus', data=df, showmeans=True, meanprops={"marker": "+",
                       "markeredgecolor": "black",
                       "markersize": "10"},ax=axs[1,2], saturation=0.75, width=0.4, palette='flare_r')
g.set(ylim=(3.5, 8.5))

g = sns.boxplot(x='Treatment_Group', y='visual_cortex', data=df, showmeans=True, meanprops={"marker": "+",
                       "markeredgecolor": "black",
                       "markersize": "10"},ax=axs[2,0], saturation=0.75, width=0.4, palette='flare_r')
g.set(ylim=(3.5, 8.5))

g = sns.boxplot(x='Treatment_Group', y='mPFC', data=df, showmeans=True, meanprops={"marker": "+",
                       "markeredgecolor": "black",
                       "markersize": "10"},ax=axs[2,1], saturation=0.75, width=0.4, palette='flare_r')
g.set(ylim=(3.5, 8.5))

g = sns.boxplot(x='Treatment_Group', y='whole_brain', data=df, showmeans=True, meanprops={"marker": "+",
                       "markeredgecolor": "black",
                       "markersize": "10"},ax=axs[2,2], saturation=0.75, width=0.4, palette='flare_r')
g.set(ylim=(3.5, 8.5))

handles, labels = g.get_legend_handles_labels()
f.legend(handles, labels, fontsize=40, loc='lower right')
#f.delaxes(axs[2,2])
#ax.set(ylim=(4, 7.6))

f.tight_layout()
```

```python
def plot_region_names (row1):
    if row1['variable'] == 'accumbens':
        return 'NAc'
    if row1['variable'] == 'visual_cortex':
        return 'visual cortex'
```

```python
df_melted = pd.melt(df, id_vars =['Rat #', 'Treatment_Group',	'Treatment_Group_#',	'Weight [kg]',	'StudyDate [date_time]',	'Time [seconds]'], value_vars =regions)
df_melted = df_melted[(df_melted.variable != 'whole_brain') & (df_melted.variable != 'frontal_cortex')] 

df_melted.loc[df_melted['variable'] == 'accumbens', 'variable'] = 'NAc'
df_melted.loc[df_melted['variable'] == 'visual_cortex', 'variable'] = 'visual cortex'


```

```python
sns.set(style="ticks", context="talk", rc={"figure.figsize":(30, 15)})

g = sns.boxplot(x='variable', y='value', data=df_melted, 
                order=['mPFC', 'OFC', 'visual cortex', 'hippocampus', 'NAc', 'striatum', 'thalamus', 'cerebellum'],
                hue='Treatment_Group', hue_order=['Veh', 'Har', 'DMT', 'Har + DMT'], showmeans=True, meanprops={"marker": "+",
                       "markeredgecolor": "black",
                       "markersize": "10"}, palette='flare_r', width=.8, boxprops={'alpha': 0.7}) 


old_len_collections = len(g.collections)
sns.stripplot(data=df_melted, x="variable", y="value", 
              order=['mPFC', 'OFC', 'visual cortex', 'hippocampus', 'NAc', 'striatum', 'thalamus', 'cerebellum'],
              hue='Treatment_Group', hue_order=['Veh', 'Har', 'DMT', 'Har + DMT'], ax=g, palette='flare_r', dodge=True) 
            
for dots in g.collections[old_len_collections:]:
    dots.set_offsets(dots.get_offsets() + np.array([-0.05, 0]))

sns.despine()

# Get the handles and labels. For this example it'll be 2 tuples of length 4 each.
handles, labels = g.get_legend_handles_labels()

# When creating the legend, only use the first two elements to effectively remove the last two.
l = plt.legend(handles[0:4], labels[0:4], bbox_to_anchor=(1.0, 1.0), loc=1, borderaxespad=0.)
for handle in l.legendHandles:
    handle.set_alpha(0.7)

g.set_ylabel('whole-brain normalized SUV', fontsize=16)
g.set_xlabel(None)

#rasterized needed for .eps to account for transparency (that comes with the confidence intervals of the regression line)
fig = g.get_figure()
fig.set_rasterized(True)
fig.savefig(f'{analysis_path}/plots_for_paper/PET_regions_plot_wb_norm.eps', dpi=300, format='eps')
fig.savefig(f'{analysis_path}/plots_for_paper/PET_regions_plot_wb_norm.png')

```
